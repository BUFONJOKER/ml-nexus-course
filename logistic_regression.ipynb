```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/credit card.csv")
```

```python
data.head()
```

```python
data.duplicated().sum()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data["type"].unique()
```

```python
data["type"] = data["type"].map({ "PAYMENT" : 0,
                                 "TRANSFER" : 1,
                                  "CASH_OUT": 2,
                                  "DEBIT" : 3,
                                  "CASH_IN": 4
                                })
```

```python
data.head()
```

```python
X = data[["type","amount","oldbalanceOrg","newbalanceOrig"]]
y = data["isFraud"]
```

```python
X.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=0.1,random_state=42)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
scaler = StandardScaler()
x_train_norm = scaler.fit_transform(x_train)
x_test_norm = scaler.transform(x_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_norm,y_train)
```

```python
y_pred = model.predict(x_test_norm)
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
results = pd.DataFrame({"Actual" : y_test , "Predicted" : y_pred})
```

```python
results.head()
```

```python
anomalies = []
count = 0
for i in range(len(x_test)):
  if y_test.iloc[i] != y_pred[i]:
    anomalies.append(f"{i} {y_test.iloc[i]}      {y_pred[i]}")
    count = count + 1
```

```python
print(f"Number of anomalies = {count} / {len(y_test)}")
```

```python
print(f"ACTUAL     PREDICTED")
for i in range(len(anomalies)):
  print(anomalies[i])
```
